```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
data = pd.read_csv("NewspaperData.csv")
print(data)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
plt.figure(figsize=(6,3))
plt.title("Box plot for Daily Sales")
plt.boxplot(data["daily"], vert = False)
plt.show()
```

```python
sns.histplot(data['daily'], kde = True,start='density',)
plt.show()
```

Obsevations
- There are no missing values
- The daily column values appear to be right-skewed
- The sunday column values also appear to be right-skewed
- There are two outliers in both daily column and also in sunday column as observed from the boxplots

```python
x= data["daily"]
y= data["sunday"]
plt.scatter(data["daily"],data["sunday"])
plt.xlim(0,max(x) +100)
plt.xlim(0,max(y) +100)
plt.show()
```

```python
data["daily"].corr(data["sunday"])
```

```python
data[["daily","sunday"]].corr()
```

Observation
- There relation between x (daily and y (sunday) is seen to be linear as seen from scatter plot
- The correlation is strong positive with Pearsons correlation of cofficient of 0.958154

```python
import statsmodels.formula.api as smf
model1 = smf.ols("sunday~daily",data = data).fit()
model1.summary()
```

```python
x = data["daily"].values
y = data["sunday"].values
plt.scatter(x, y, color = "m", marker = "o", s = 30)
b0 = 13.84
b1 = 1.33
y_hat = b0 + b1*x
plt.plot(x, y_hat, color = "g")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
sns.regplot(x="daily", y="sunday", data=data)
plt.xlim([0,1250])
plt.show()
```

```python
#Predict sunday sales for 200 and 300 and 1500 daily circulation
newdata=pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
pred = model1.predict(data["daily"])
pred
```

```python
data["Y_hat"] = pred
data
```

```python
data["residuals"]= data["sunday"]-data["Y_hat"]
data
```

```python
mse = np.mean((data["daily"]-data["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RSME: ",rmse)
```
